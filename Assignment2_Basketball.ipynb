```python
%matplotlib inline

import numpy as np   # Import libraries for the diagram.
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt  


from matplotlib.figure import Figure

from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from matplotlib.patches import Arc
from matplotlib.patches import Patch
```

```python
# This following function will replicate an NBA basketball court.

# The reference used for this assignment will be from the following website to replicate the dimensions.

# https://proformancehoops.com/basketball-court-dimensions/

def nba_basketball_court():
    fig1=plt.figure() 
    ax=fig1.add_subplot(111)

    
    # 1. First, draw the boundaries of the court.  The dimensions are 94 feet by 50 feet.  Then add two keys on both ends of the 
    # court.  Each of 19 feet by 16 feet.

    outer_Rec=matplotlib.patches.Rectangle((0,0),94,50,angle=0.0,edgecolor='black',facecolor='beige')
    
    left_key=matplotlib.patches.Rectangle((0,17),19,16,angle=0.0,edgecolor='red',facecolor='none')
    inner_left_key=matplotlib.patches.Rectangle((0,19),19,12,angle=0.0,edgecolor='red',facecolor='none')
    
    right_key=matplotlib.patches.Rectangle((75,17),19,16,angle=0.0,edgecolor='red',facecolor='none')
    inner_right_key=matplotlib.patches.Rectangle((75,19),19,12,angle=0.0,edgecolor='red',facecolor='none')
    ax.add_patch(outer_Rec)
    ax.add_patch(left_key)
    ax.add_patch(right_key)
    ax.add_patch(inner_left_key)
    ax.add_patch(inner_right_key)
    
    # 2. Second, add the lines within the court.  The division line is drawn in the middle, 47 feet long.  The widths of the lines
    # will be negligible measuring only two inches.  Moreover, the lines around the half court will be slightly extended
    # from its original measurement of 3 inches.

    divLine=plt.vlines(47,0,50,colors='red',linestyles='solid')
    
    # Vertical Lines

    vlin1=plt.vlines(28,0,3,colors='black',linestyles='solid')
    vlin2=plt.vlines(28,47,50,colors='black',linestyles='solid')
    vlin3=plt.vlines(66,0,3,colors='black',linestyles='solid')
    vlin4=plt.vlines(66,47,50,colors='black',linestyles='solid')
    
    out_lin1=plt.vlines(43,-3,0,colors='black',linestyles='solid')
    out_lin2=plt.vlines(51,-3,0,colors='black',linestyles='solid')
    
    # Horizontal Lines
    
    hlin1=plt.hlines(14,0,1.5,colors='black',linestyles='solid')
    hlin2=plt.hlines(14,92.5,94,colors='black',linestyles='solid')
    hlin3=plt.hlines(36,0,1.5,colors='black',linestyles='solid')
    hlin4=plt.hlines(36,92.5,94,colors='black',linestyles='solid')
    
    # Vertical lines around the left key.
    
    v_leftkey1=plt.vlines(7,33,34,colors='black',linestyles='solid')
    v_leftkeydown1=plt.vlines(7,16,17,colors='black',linestyles='solid')
    
    v_leftkey2=plt.vlines(8,33,34,colors='black',linestyles='solid')
    v_leftkeydown2=plt.vlines(8,16,17,colors='black',linestyles='solid')
    
    v_leftkey3=plt.vlines(11,33,34,colors='black',linestyles='solid')
    v_leftkeydown3=plt.vlines(11,16,17,colors='black',linestyles='solid')
    
    v_leftkey2=plt.vlines(14,33,34,colors='black',linestyles='solid')
    v_leftkeydown2=plt.vlines(14,16,17,colors='black',linestyles='solid')
    
    # Vertical lines around the right key.
    
    v_rightkey1=plt.vlines(80,33,34,colors='black',linestyles='solid')
    v_rightkeydown1=plt.vlines(80,16,17,colors='black',linestyles='solid')
    
    v_rightkey2=plt.vlines(83,33,34,colors='black',linestyles='solid')
    v_rightkeydown2=plt.vlines(83,16,17,colors='black',linestyles='solid')
    
    v_rightkey3=plt.vlines(86,33,34,colors='black',linestyles='solid')
    v_rightkeydown3=plt.vlines(86,16,17,colors='black',linestyles='solid')
    
    v_rightkey2=plt.vlines(87,33,34,colors='black',linestyles='solid')
    v_rightkeydown2=plt.vlines(87,16,17,colors='black',linestyles='solid')
    
    # Add lines connecting to the arcs.
    
    upline_left=plt.hlines(47,0,14,colors='black',linestyles='solid')
    downline_left=plt.hlines(3,0,14,colors='black',linestyles='solid')
    
    upline_right=plt.hlines(47,80,94,colors='black',linestyles='solid')
    downline_right=plt.hlines(3,80,94,colors='black',linestyles='solid')
    
    #3. Add circles and the tip-off circles on each helf-point court.
    
    small_cen=matplotlib.patches.Circle((47,25),radius=2,edgecolor='red',facecolor='none')
    big_cen=matplotlib.patches.Circle((47,25),radius=6,edgecolor='red', facecolor='none')
    
    ax.add_patch(small_cen)
    ax.add_patch(big_cen)
    
    left_solid_tipoff=matplotlib.patches.Arc((19,25),12,12,angle=0.0,theta1=-90.0,theta2=90.0,edgecolor='red',facecolor='none')
    left_dotted_tipoff=matplotlib.patches.Arc((19,25),12,12,angle=0.0,theta1=90.0,theta2=270.0,edgecolor='red',facecolor='none',linestyle='--')
    
    right_solid_tipoff=matplotlib.patches.Arc((75,25),12,12,angle=0.0,theta1=90.0,theta2=270.0,edgecolor='red',facecolor='none')
    right_dotted_tipoff=matplotlib.patches.Arc((75,25),12,12,angle=0.0,theta1=-90.0,theta2=90.0,edgecolor='red',facecolor='none',linestyle='--')
    
    ax.add_patch(left_solid_tipoff)
    ax.add_patch(left_dotted_tipoff)
    
    ax.add_patch(right_solid_tipoff)
    ax.add_patch(right_dotted_tipoff)
    
    #4. Create the three-point arcs.  The centers of the three-point arcs are 23.75 ft from the center of the basket each.
    
    three_point_arc_left=matplotlib.patches.Arc((4.25,25),48.5,48,angle=0.0,theta1=-65.95,theta2=65.95,edgecolor='black',facecolor='none')
    ax.add_patch(three_point_arc_left)
    
    three_point_arc_right=matplotlib.patches.Arc((89.75,25),48.5,48,angle=0.0,theta1=114.05,theta2=245.95,edgecolor='black',facecolor='none')
    ax.add_patch(three_point_arc_right)
    
    #5. Incorporate the basketball nets in a similar method to creating the three-point arcs.
    
    left_backboard=plt.vlines(4,22,28,colors='orange',linestyles='solid')
    right_backboard=plt.vlines(90,22,28,colors='orange',linestyles='solid')
 
    left_basket=matplotlib.patches.Circle((4.75,25),radius=0.75,edgecolor='orange',facecolor='orange')
    right_basket=matplotlib.patches.Circle((89.25,25),radius=0.75,edgecolor='orange',facecolor='orange')
    
    ax.add_patch(left_basket)
    ax.add_patch(right_basket)
    
    #6. Finally, add the foul lines just outside the nets.
    
    foul_up_left=plt.hlines(29,4,5.25,colors='blue',linestyles='solid')
    foul_down_left=plt.hlines(21,4,5.25,colors='blue',linestyles='solid')
    
    left_foul_arc=matplotlib.patches.Arc((4.75,25),8,8,angle=0.0,theta1=-90.0,theta2=90.0,edgecolor='blue',facecolor='none')
    ax.add_patch(left_foul_arc)
    
    foul_up_right=plt.hlines(29,88.75,90,colors='blue',linestyles='solid')
    foul_down_right=plt.hlines(21,88.75,90,colors='blue',linestyles='solid')
    
    right_foul_arc=matplotlib.patches.Arc((89.25,25),8,8,angle=0.0,theta1=90.0,theta2=270.0,edgecolor='blue',facecolor='none')
    ax.add_patch(right_foul_arc)
   
       
    # Save the court as a png file.
    
 
    plt.savefig("nba_court.png")
    
    plt.show()
```
